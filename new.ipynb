```python
url_cn = "https://api.covid19api.com/country/korea-north/status/confirmed/live"
```

```python
df_cn = pd.read_json(url_cn)
```

```python
df_cn
```

```python
df_cn.set_index("Date")["Cases"]
```

```python
casos_esp = df_esp.set_index("Date")["Cases"]
casos_ale = df_ale.set_index("Date")["Cases"]
casos_cn = df_cn.set_index("Date")["Cases"]
```

```python
vs = pd.concat([casos_esp,casos_ale,casos_cn],axis=1)
```

```python
vs
```

```python
vs.columns = ["España","Alemania","Corea del Norte"]
```

```python
vs
```

```python
vs.plot(title="Gráfico de comparación entre países")
```

# DATOS

```python
vs.to_csv("esvsit.csv")
```
